# Importing libraries and dependencies

```python
import numpy as np
import tensorflow as tf
from tpot import TPOTClassifier
import pandas as pd
```

# Dataset Loading

```python
data = pd.read_csv("../data/wine.csv")
```

```python
data.drop(["Unnamed: 0"],inplace = True,axis = 1)
```

```python
# Shuffling
data = data.sample(frac=1)
data
```

# Train and Test Split

```python
split_length = int(data.shape[0] * 0.8)  # 141
```

```python
# train and test
train_data = data.iloc[:split_length]
test_data = data.iloc[split_length:]
```

# Selecting Features and Labels

```python
features = data.columns[1:]
label = data.columns[0:1]
```

```python
features
```

```python
label
```

# Initializing Classifier

```python
tpot = TPOTClassifier(verbosity=2, max_time_mins=2)
```

# Search for the best model

```python
# Feed the classifier with training data.

tpot.fit(train_data[features], train_data[label])
print(tpot.score(test_data[features], test_data[label]))
```

# Exporting Model

```python
tpot.export('tpot_wine_pipeline.py')
```
